```python
from VecUtils import ImgVectorize, TextVectorize
import pandas as pd
df = pd.read_csv('~/Desktop/Ops Con/ChangingRoom/Coding/Data/sample_data.csv')
df_100 = df.sample(n=100).reset_index()
img_vec_ls = []
txt_vec_ls = []

for index, row in df_100.iterrows():
    img_vec = ImgVectorize(row['Image URL'])   
    img_vec_ls.append(img_vec.get_vec())

    sentence = row.drop(['index', 'Code', 'Price', 'In Stock', 'brand_code', 'item_number', 'URL', 'Image URL', 'Collection']).dropna()
    concat_str = ""
    for i in sentence:
        concat_str += i +". "
    txt_vec = TextVectorize(concat_str)
    txt_vec_ls.append(txt_vec.get_vec())

df_100['ImgVec'] = img_vec_ls
df_100['TxtVec'] = txt_vec_ls
```
